## **Checking For Stationarity**

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(dataset):
     dftest = adfuller(dataset, autolag = 'AIC')
     print("1. ADF : ",dftest[0])
     print("2. P-Value : ", dftest[1])
     print("3. Num Of Lags : ", dftest[2])
     print("4. Num Of Observations Used For ADF Regression:",      dftest[3])
     print("5. Critical Values :")
     for key, val in dftest[4].items():
         print("\t",key, ": ", val)
adf_test(df['Parboiled rice'])
```

## **SPLIT THE DATA**

```python
print(df.shape)
train=df.iloc[:-30]
test=df.iloc[-30:]
print(train.shape,test.shape)
```

## **SUMMARY**

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train['Parboiled rice'], order=(2,0,2))
model = model.fit()
print(model.summary())
```

## **PLOT**

```python
start=len(train)
end=len(train)+len(test)-1
pred=model.predict(start=start,end=end,typ='levels').rename('ARIMA Predictions')
pred.plot(legend=True)
test['Parboiled rice'].plot(legend=True)
```

## **ERROR**

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
m=test['Parboiled rice'].mean()
print('MEAN = ',m)
rmse=sqrt(mean_squared_error(pred,test['Parboiled rice']))
print('ROOT MEAN SQUARE ERROR = ',rmse)
avgerr = (rmse / m) * 100
print('AVERAGE ERROR = ',avgerr)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

# Calculate R2 score
r2 = r2_score(test['Parboiled rice'], pred)
print('R2 SCORE =', r2)

# Calculate MAE
mae = mean_absolute_error(test['Parboiled rice'],pred )
print('MEAN ABSOLUTE ERROR =', mae)
```
